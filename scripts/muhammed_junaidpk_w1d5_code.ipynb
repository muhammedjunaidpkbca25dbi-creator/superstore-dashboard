data cleaning

```python
print("duplicate rows:", df.duplicated().sum())
dupes = df[df.duplicated(subset=['Order ID'], keep=False)]
print(dupes.shape)
dupes.head()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
df["Shipping_days"] = (df["Ship Date"] - df["Order Date"]).dt.days
```

```python
print(df.groupby("Region")["Shipping_days"].mean().round(1))
```

```python
df["Profit_Margin"] = df["Profit"] / df["Sales"]*100
print(df.groupby("Sub-Category")["Profit_Margin"].mean().sort_values().head(10).round(2))
```

```python
df["order_Year"] = df["Order Date"].dt.year
```

```python
df.rename(columns={"Sub-Category": "sub_category",
                    "Ship Mode": "ship_Mode",
                    "Customer Name": "customer_Name",
                    "Segment": "segment",
                    "Country": "country",
                    "City": "city",
                    "Customer ID": "customer_ID",
                    'Row ID': 'row_ID',
                    "Order ID": "order_ID",
                    "Order Date": "order_date",
                    "Ship Date": "ship_date",
                    'Country/Region': 'country_Region',
                    'State/Province': 'state_Province',
                    'Postal Code': 'postal_Code',
                    'Region': 'region',
                    'Product ID': 'product_ID',
                    'Category': 'category',
                    'Sales': 'sales',
                    'Quantity': 'quantity',
                    'Discount': 'discount',
                    'Profit': 'profit',
                    'Shipping_days': 'shipping_days',
                    'Profit_Margin': 'profit_Margin',
                    'order_Year': 'order_year'
                    }, inplace=True)
```
